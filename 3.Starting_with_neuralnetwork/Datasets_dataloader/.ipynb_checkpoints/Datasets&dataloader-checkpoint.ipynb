```python
import torch 
from torchvision import datasets
from torchvision.transforms import ToTensor
```

```python
training_datsets = datasets.FashionMNIST (
    root = "DATA" ,
    download = True , 
    train = True , 
    transform = ToTensor
)

test_datsets = datasets.FashionMNIST (
    root = "DATA" ,
    download = True , 
    train = False , 
    transform = ToTensor
)
```
